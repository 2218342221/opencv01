## 采样
 

```python
# %%
import cv2
import numpy as np


# %% 显示图片
def show(o):
    cv2.imshow("show", o)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


# %%
lena = cv2.imread("resources/lena.jpg")

# %% 转灰度
lena_gray = cv2.imread("resources/street_2.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# 放大 缩小

lena_up = cv2.pyrUp(lena)
show(lena_up)
print(lena_up.shape)

lena_down = cv2.pyrDown(lena)
show(lena_down)
print(lena_down.shape)
```

拉普拉斯金字塔 先放大，再缩小， 与原图的差  --》轮廓

```python
lena_down_up = cv2.pyrUp(lena_down)
show(lena-lena_down_up)
```
